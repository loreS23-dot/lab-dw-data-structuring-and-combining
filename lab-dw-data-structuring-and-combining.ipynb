## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

df1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")



rename_map = {'ST': 'State', 'GENDER': 'Gender'}
df1 = df1.rename(columns=rename_map)
df2 = df2.rename(columns=rename_map)
df3 = df3.rename(columns=rename_map)

combined = pd.concat([df1, df2, df3], ignore_index=True)

print("Filas df1:", len(df1))
print("Filas df2:", len(df2))
print("Filas df3:", len(df3))
print("Filas total tras concat:", len(combined))


#limpiamos costumer lifetime value
combined["Customer Lifetime Value"] = (combined["Customer Lifetime Value"].astype(str).str.replace("%", "", regex=False).str.replace(",", "", regex=False).astype(float) / 100)
#limpiamos number of open complaints
combined["Number of Open Complaints"] = (combined["Number of Open Complaints"].astype(str).str.split("/").str[1])
combined["Number of Open Complaints"] = pd.to_numeric(combined["Number of Open Complaints"],errors="coerce")
#Sustutuimos los valores nulos de la columna por "Unknown"
combined["Gender"] = combined["Gender"].fillna("Unknown")
#Estandarizamos los valores de la columna Education
combined["Education"] = combined["Education"].replace({
    "bacherlors": "Bachelor",
    "Bacherlors": "Bachelor",
    "Bachelors": "Bachelor"
})


combined.head(5)




```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
print("Filas df:", len(df))

df["revenue"] = (df["monthly_premium_auto"] * df["months_since_policy_inception"]- df["total_claim_amount"])

pivot_revenue = df.pivot_table(
    index="sales_channel",    # canal de venta
    values="revenue",         # nuestra métrica nueva
    aggfunc="sum"             # suma total por canal
)

pivot_revenue = pivot_revenue.round(2)

print(pivot_revenue)


print(pivot_revenue)

pivot_clv = df.pivot_table(
    index="gender",                         # filas = género
    columns="education",                    # columnas = nivel educativo
    values="customer_lifetime_value",       # métrica
    aggfunc="mean"                          # promedio
)

# Redondeamos a 2 decimales para que sea más legible
pivot_clv = pivot_clv.round(2)

print(pivot_clv)


```
